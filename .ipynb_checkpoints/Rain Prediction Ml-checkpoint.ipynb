**import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import resample
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import pickle
```

**Data Collection and Processing**

```python
data = pd.read_csv("weather_data.csv")
# remove extra spaces in each columnabs
data.columns.str.strip()
print("Data Info:")
data.info()
```

```python
# drop the unnessasary column date
data = data.drop(columns=['date'])
data.info()
data.head()
```

```python
# check for number of missing values
data.isnull().sum()
```

```python
def handle_missing_values(data):
    for col in ['avg_temperature', 'humidity', 'avg_wind_speed', 'cloud_cover', 'pressure']:
        data[col] = pd.to_numeric(data[col], errors='coerce')
        if data[col].isnull().sum() > 0:
            data[col].fillna(data[col].median(), inplace=True)
    return data

handle_missing_values(data)
```

```python
# map Rain to 1 and No Rain to 0
data['rain_or_not'] = data['rain_or_not'].map({"Rain" : 1, "No Rain" : 0})
data['rain_or_not'].unique()
```

```python
data["rain_or_not"].value_counts()
```

```python
# down sampling will apply from here because there are too much of 1's for this rain_or_not column

# separate majority and minority
df_majority = data[data["rain_or_not"] == 1]
df_minority = data[data["rain_or_not"] == 0]
print(df_majority.shape)
print(df_minority.shape)
```

```python
# downsampling majority class to match minority count
df_majority_downsampled = resample(df_majority, replace=False, n_samples=len(df_minority), random_state=42)
print(df_majority_downsampled.shape)
```

```python
# downsapled ddata
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
df_downsampled.shape
```

```python
df_downsampled.head()
```

```python
# shuffling final dataframe
df_downsampled = df_downsampled.sample(frac=1, random_state=42).reset_index(drop= True)
df_downsampled.head()
```

```python
df_downsampled["rain_or_not"].value_counts()
# this have 113 number of Rain and No Rain rows
```

```python
# splitting the data into training data and test data
X = df_downsampled.drop(columns=["rain_or_not"])
Y = df_downsampled["rain_or_not"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
rf_model = RandomForestClassifier(random_state=42)

param_grid_rf = {
    "n_estimators" : [50, 100, 200],
    "max_features" : ["sqrt", "log2"],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split" : [2,5,10],
    "min_samples_leaf" : [1, 2, 4]
}
```

```python
# Hypertuning using GridSearchCV
grid_search_rf = GridSearchCV(estimator=rf_model, param_grid=param_grid_rf, cv=5, n_jobs=-1, verbose=2)
grid_search_rf.fit(X_train, Y_train)
```

```python
# who gives the highest aaccuracy
best_rf_model = grid_search_rf.best_estimator_
print("best parameters form Random Forest:", grid_search_rf.best_params_)
```

**Model Evaluation**

```python
cv_scores = cross_val_score(best_rf_model, X_train, Y_train, cv=5)
print("Cross-validation scores:", cv_scores)
print("Mean cross-validation score:", np.mean(cv_scores))
```
